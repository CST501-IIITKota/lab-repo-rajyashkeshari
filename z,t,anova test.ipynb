```python
null_mean = 0.9  # hypothesis
```

```python
from statsmodels.stats.weightstats import ztest
zscore, pvalue = ztest(population, value=null_mean, alternative = 'larger')
```

```python
zscore
```

```python
if pvalue<0.05:
    print('Reject the null hypothesis')
else:
    print('Failure to reject the null hypothesis')
```

```python
#z critical approach

#A school calculated iq scores of 50 students, the average iq  turned out to be 100. the mean of population iq is 90 
#and the std deviation 16. \
#state whether the claim by school 
#that IQ increases if student study more than avg of school with 5% of significance level

#null hypothesis : mean_iq = 90
#Alternate: mean_iq > 90
```

```python
import scipy.stats as stats
sample_mean = 100
population_mean = 90
standarad_deviation = 16
sample_size = 50
alpha = 0.05
```

```python
zscore = (sample_mean - population_mean)/(standarad_deviation/(np.sqrt(sample_size)))
```

```python
zscore
```

```python
zcritical = stats.norm.ppf(1-alpha)
```

```python
zcritical
```

```python
if zscore > zcritical:
    print("reject the null hypothesis")
else:
    print("Failure to reject null hypothesis")
```

```python
#using bscipy.stats pvalue approcah
p_value = 1 - stats.norm.cdf(zscore)
if p_value<=0.05:
    print("Reject the null hypothesis")
else:
    print("Failure to reject the null hypothesis")
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
data[["sex", "smoker"]]
```

```python
#cross tab : summarising of data
data_lab = pd.crosstab(data["sex"], data["smoker"])
```

```python
data_lab
```

```python
observed_value = data_lab.values
```

```python
observed_value
```

```python
import scipy.stats as stats
stats_test, pvalue, dof, expected_values = stats.chi2_contingency(observed_value)
```

```python
dof = 1
expected_values
```

```python
from scipy.stats import chi2
```

```python
for i in zip(observed_value, expected_values):
    print(i)
```

```python
chisquare_test = sum([(o-e)**2/e for o, e in zip(observed_value, expected_values)])
```

```python
chisquare_test
```

```python
chisquare_stats = chisquare_test[0] + chisquare_test[1]
```

```python
chisquare_stats
```

```python
alpha = 0.05
```

```python
chi2_critical = chi2.ppf(1-alpha, dof)
```

```python
chi2_critical
```

```python
if chisquare_stats >= chi2_critical:
    print("Reject the null hypothesis, there is a relationship between tow categorical variables")
else:
    print("fail to reject null hypothesis, no relationship")
```

```python
#pvalue method
p_value = 1-chi2.cdf(x = chisquare_stats, df = dof)
```

```python
if p_value <= 0.05:
    print("Reject the null hypothesis, there is a relationship between tow categorical variables")
else:
    print("fail to reject null hypothesis, no relationship")
```
