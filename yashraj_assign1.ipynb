3. Outlier Detection and Handling

```python
plt.boxplot(data['price'], vert=False)
plt.title("Box Plot of Price Distribution")
plt.xlabel("Price")
plt.show()
```

```python
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3 - Q1

print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR (Q3 - Q1): {IQR}")
```

```python
 # 3. Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")
```

```python
# 4. Create new DataFrame excluding outliers
data_cleaned = data[(data['price'] >= lower_bound) & (data['price'] <= upper_bound)]

print("\nData without outliers:")
print(data_cleaned)
```

**Problem Statement 3: Sampling and the Central Limit Theorem (CLT)**

Objective: To demonstrate the Central Limit Theorem by drawing samples from a non-normally distributed population.

1. Population Distribution:

```python
plt.figure(figsize=(10,5))
sns.histplot(data['calculated_host_listings_count'], bins=50, kde=False)
plt.title("Distribution of Host Listings Count (Population)")
plt.xlabel("Number of Listings per Host")
plt.ylabel("Frequency")
plt.yscale("log")   # optional: makes skew easier to see
plt.show()
```

```python
sample_means = []

for i in range(1000):
    sample = data['calculated_host_listings_count'].sample(n=50, replace=True, random_state=None)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)

sample_means = np.array(sample_means)

print("First 10 sample means:", sample_means[:10])
print("Average of sample means:", sample_means.mean())
```

```python
plt.figure(figsize=(10,5))
sns.histplot(sample_means, bins=30, kde=True)
plt.title("Sampling Distribution of Sample Means (n=50, 1000 samples)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
print("Mean of Sample Means:", sample_means.mean())
```

Explain how your findings demonstrate the Central Limit Theorem.

If you repeatedly take random samples of size n from any population and compute their means, the distribution of those sample means will approach a normal distribution, regardless of the original population’s shape.
